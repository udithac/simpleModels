# Model Generation - Rabbits vs Foxes

## Generate a program to model the populations of rabbits and foxes in a habitat.

### Information 
#### 1. Foxes eat rabbits
#### 2. Foxes die of staration, if they don't eat enough rabbits.
#### 3. Rabbits and foxes both reproduce, but rabbits reproduce faster.
#### 4. Rabbits and foxes both die of other causes.

### Solution
#### Understanding the problem
- Foxes eat rabbits: Means fox population depends on availability of rabbits
- Rabbits reproduce faster than foxes: Rabbits population will grow in a faster rate than foxes
- Both die by other factors: Need to consider a motality rate for both

### 1. Init variables
##### Let's assume

```python
# Population
rabbits = 1000
foxes = 20

# Reproduction rate
rabbit_birth_rate = .1
fox_birth_rate = .05

# Motality rates
rabbit_death_rate = .02
fox_death_rate = .01

# Number of time steps to simulation 
simulation_days = 365
```

#### 2. Iterate through time steps
##### Need to iterate through simulation days via a loop

```python
# Rabbit Re=production
rabbits += rabbits * rabbit_birth_rate
# Fox reproduction
foxes += foxes * fox_birth_rate


# Simulate fox production
# Let's define, foxes eat rabbits rate
prediction_rate = .02 # each day

# rabbit population
rabbits -= foxes * prediction_rate * rabbits

# Motality rate application
rabbits -= rabbits * rabbit_death_rate

foxes -= foxes * fox_death_rate
```

```python
rabbits = 1000
foxes = 20

rabbit_birth_rate = 0.1
fox_birth_rate = 0.05
rabbit_death_rate = 0.02
fox_death_rate = 0.02
predation_rate = 0.02

simulation_days = 365

for day in range(simulation_days):
  rabbits += rabbits * rabbit_birth_rate
  foxes += foxes * fox_birth_rate
  rabbits -= foxes * predation_rate * rabbits
  rabbits -= rabbits * rabbit_death_rate
  foxes -= foxes * fox_death_rate

  # Print results (optional)
  print(f"Day {day+1}: Rabbits = {rabbits:.2f}, Foxes = {foxes:.2f}")
```

The reason we're getting "nan" (Not a Number) for the rabbit population after a certain point in your simulation is because the rabbit population has become extinct. 

Here's what's happening:

Fox Predation: The code includes a predation rate where foxes consume rabbits.
Rabbit Mortality: Also have a separate mortality rate for rabbits that reduces their population naturally.
Unbalanced Rates: If the combined effect of predation and mortality is greater than the rabbit birth rate, the rabbit population will eventually decline to zero.

Explanation of the Output:
The fox population keeps increasing because it's still consuming the remaining rabbits and reproducing at its own rate.
As the rabbit population reaches very low numbers (close to zero), even a small predation rate can significantly impact the remaining rabbits, leading to a population crash.

How to Fix It:

There are a few ways to address this and make the simulation more realistic:

Adjust Rabbit Birth Rate: Increase the rabbit birth rate so it's slightly higher than the combined effect of predation and mortality. This allows the rabbit population to sustain itself despite predation pressure.
Carrying Capacity: Introduce a concept of carrying capacity for rabbits. This represents the maximum number of rabbits the habitat can support due to resource limitations. Can modify the rabbit birth rate based on the current rabbit population and the carrying capacity. As the rabbit population approaches the carrying capacity, the birth rate can decrease, simulating resource limitations.

Environmental Changes: Introduce random variations in birth and death rates to simulate unpredictable environmental factors. This can add complexity and make the simulation more dynamic.
